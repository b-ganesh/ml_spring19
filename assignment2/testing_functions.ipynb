#### for example, if I wanted to see which rows had outliers for the age variable, I could do so

```python
outlier_vars = ['age']
outliers = assignment2_functions.identify_ol(df, outlier_vars)
outliers.head()
```

```python
all_outliers = assignment2_functions.identify_ol(df)
data_without_outliers = pd.merge(df, all_outliers, how='outer', indicator=True)
data_without_outliers = data_without_outliers.loc[data_without_outliers._merge == 'left_only']
data_without_outliers.drop(labels='_merge', axis=1, inplace=True)
data_without_outliers
```

#### Finally, we see how different the distributions look if outliers are removed

```python
assignment2_functions.histograms(data_without_outliers)
```
