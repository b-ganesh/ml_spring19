## Part 2: Process Data

#### Below, I process the data in the columns that have NA values, and I replace the NAs with the median of the values in that column. I then summarize NAs for the columns again to make sure that there are no remaining columns with NA values

```python
df = assignment2_functions.pre_process(df)
na_summary = assignment2_functions.na_summary(df)
na_summary
```

## Part 3: Generate Features

#### Based on my analysis of outliers above, I decided to discretize age, MonthlyIncome, and NumberOfOpenCreditLinesAndLoans, because their distributions seem to be the most affected by the presence of outliers.

```python
vars_to_discretize = ['age', 'MonthlyIncome', 'NumberOfOpenCreditLinesAndLoans']
df = assignment2_functions.discretize(df, vars_to_discretize)
df.head()
```

#### I then made dummy values for zipcode, which appears to be the only categorical variable in the dataset. I also made dummy variables for the three variables I just discretized.

```python
df = assignment2_functions.categorize(df, ['zipcode', 'MonthlyIncome_discrete', 'NumberOfOpenCreditLinesAndLoans_discrete', 'age_discrete'])
df.head()
```

## Part 4: Selecting Features and Training a Model

#### Now that I've created discretized and dummy variables, I drop the original columns from the data. I also drop highly collinear columns (NumberOfTime30-59DaysPastDueNotWorse and NumberOfTime60-89DaysPastDueNotWorse)

```python
selected_features = [col for col in df if col.startswith('zipcode') or col not in ['PersonID', 'SeriousDlqin2yrs', 'zipcode', 'age', 'MonthlyIncome', 'NumberOfTime30-59DaysPastDueNotWorse', 'NumberOfTime60-89DaysPastDueNotWorse','NumberOfOpenCreditLinesAndLoans', 'NumberOfOpenCreditLinesAndLoans_discrete', 'MonthlyIncome_discrete', 'age_discrete']]
selected_y = 'SeriousDlqin2yrs'
```

#### Now I create train, test splits using these selected features 

```python
x_train, x_test, y_train, y_test = assignment2_functions.split_data(df, selected_features, selected_y, test_size=0.3)
```

#### Using these train/test splits, I build a decision tree classifier model. Without specifying a max_depth, the accuracy of the model is 80.7%. This means that the model predicts the y-variable serious delinquency in 2 years correctly 80.7% of the time.

```python
y_pred, dec_tree = assignment2_functions.build_classifier(x_train, x_test, y_train, y_test, max_depth=None)
accuracy = assignment2_functions.evaluate_classifier(y_test, y_pred)
accuracy
```

#### Finally, I visualize the decision tree, setting a max depth of 5 so that the tree can be easily viewed and interpreted

```python
y_pred, dec_tree = assignment2_functions.build_classifier(x_train, x_test, y_train, y_test, max_depth=5)
class_names = ['Seriously delinquent in 2 years', 'Not seriously delinquent in 2 years']
export_graphviz(
dec_tree, feature_names=x_train.columns, class_names=class_names, rounded=True, filled=True, out_file="output_files/tree.dot")
graphviz.render('dot', 'png', "tree.dot")
graphviz.Source.from_file("output_files/tree.dot")
```
